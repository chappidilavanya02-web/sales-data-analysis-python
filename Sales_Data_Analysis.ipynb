```python
##-- step 5 data visualization
##--1. what is the best year of sales?
year_group = ds.groupby("year")["item_id"].count()
year_group
```

```python
## --2. what was the best month of sales?
month_group = ds.groupby("month")["item_id"].count()
month_group

mp.plot(month_group.index,month_group.values, color="cyan")
mp.xlabel("months")
mp.ylabel("No.of sales")
mp.title("number of sales in each month")


```

```python
##-- 3.group by year and calculate total sales?
totalsales_group = ds.groupby("year")["amount"].sum()
totalsales_group
```

```python
mp.bar(totalsales_group.index, totalsales_group.values, color= "grey")
mp.plot(totalsales_group.index, totalsales_group.values, color="brown")
mp.xlabel("Years")
mp.title("total sales happend in each year")
```

```python
##--4. What brand sold the most in 2015 to 2018?
brand_15to18 = ds[(ds["year"]>=2015)&(ds["year"]<=2018)]
brand_sales = brand_15to18.groupby("brand")["amount"].sum().nlargest(10)
brand_sales
```

```python
##--5.create different plots for each year 2015, 2016,2017 what brand sold the most?
#filtering data for 2016,2017,2018
ds_2016= ds[ds["year"]==2016]
ds_2017= ds[ds["year"]==2017]
ds_2018= ds[ds["year"]==2018]

#creating groups of brand
brand_group_2016 = ds_2016.groupby("brand")["amount"].sum().nlargest(5)
brand_group_2017 = ds_2017.groupby("brand")["amount"].sum().nlargest(5)
brand_group_2018 = ds_2018.groupby("brand")["amount"].sum().nlargest(5)

#creating subpolots 
fig, axs = mp.subplots(1,3, figsize=(15,4))

# plot for first subplot
axs[0].stackplot(brand_group_2016.index, brand_group_2016.values, color="blue")
axs[0].set_title('Top 5 brands in 2016')

# plot for second subplot
axs[1].stackplot(brand_group_2017.index, brand_group_2017.values, color="purple")
axs[1].set_title('Top 5 brands in 2017')

# plot for second subplot
axs[2].stackplot(brand_group_2018.index, brand_group_2018.values, color="lightseagreen")
axs[2].set_title('Top 5 brands in 2018')
```

```python
##-- 6.what product by category sold most?
category_sales = ds.groupby("category")["amount"].sum()
category_sales
```

```python
mp.bar(category_sales.index, category_sales.values, color="orange")
mp.xlabel("category")
mp.ylabel("total sales")
mp.xticks(rotation=75)
mp.title(" top categories based on total sales")
```

```python
##--7. what product by brand name sold the least between 2015 to 2018?
brand_15to18 = ds[(ds["year"]>=2015)&(ds["year"]<=2018)]
brand_sales = brand_15to18.groupby("brand")["amount"].sum().nsmallest (10)
brand_sales
```

```python
mp.bar(brand_sales.index, brand_sales.values, color= "blue")
mp.xlabel("brand")
mp.ylabel("number of sales")
mp.xticks(rotation= 90)
mp.title("brand name sold the least between 2015 to 2018")
```

```python
##--8.the distribution of rating?
rating_counts= ds.groupby("rating")["user_id"].count().sort_index()
rating_counts
```

```python
fig, ax = mp.subplots()

# Bar plot
ax.bar(
    rating_counts.index,
    rating_counts.values,
    color="purple",
    label="Ratings Count"
)

# Line plot
ax.plot(
    rating_counts.index,
    rating_counts.values,
    marker="o",
    color="blue",
    label="Trend"
)

ax.set_xlabel("Rating")
ax.set_ylabel("Number of Users")
ax.set_title("Rating Distribution with Trend Line")
ax.legend()

mp.show()
```

```python
##-- 9. what is the most rated brand name in 2015 to 2018?
brand_15to18 = ds[(ds["year"]>=2015)&(ds["year"]<=2018)]

brand_rating_counts= brand_15to18.groupby("brand")["rating"].count().nlargest(10)
brand_rating_counts
```
